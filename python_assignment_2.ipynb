Create the below pattern using nested for loop in Python

*
* *
* * *
* * * *
* * * * *
* * * *
* * *
* *
*

```python
n=int(input("Enter number of rows: "))
      
for i in range(n):    
    if i < n/2:
        for j in range(i+1):
            print('* ',end='')
    else:
        for k in range(n-i):
            print('* ',end='')
    print()
```

Write a Python program to reverse a word after accepting the input from the user
Sample Output:
input word: ineuron
output word: norueni

```python
a = input("input word: ")
b = a[::-1]
print("output word: {}".format(b))
```
